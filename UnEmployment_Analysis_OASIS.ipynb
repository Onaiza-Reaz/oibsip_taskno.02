# Average of Top 10 Unemployed Ragion

```python
top_10=df.groupby("Region")[" Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False)
```

```python
top_10.head(10).sort_values(ascending=False)
```

```python
sns.barplot(x=df["Region"],y=df[" Estimated Unemployment Rate (%)"],data=df)
plt.title("Unemployment with respect to Region")
plt.xticks(rotation=90)
plt.show()
```

# Average of Top 10 Employed Region

# Unemployed Region According to Dates

```python
top_10_R=df.nlargest(10," Estimated Unemployment Rate (%)")[[" Estimated Unemployment Rate (%)"," Date","Region"]].set_index(" Estimated Unemployment Rate (%)")
top_10_R
```

```python
sns.barplot(x=top_10_R.index, y="Region",data=top_10_R, hue=" Date",dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```

# Employed Region According to Dates

```python
top_10_I=df.nlargest(10," Estimated Employed")[[" Estimated Employed"," Date","Region"]].set_index(" Estimated Employed")
top_10_I
```

```python
sns.barplot(x=top_10_I.index, y="Region",data=top_10_I, hue=" Date",dodge=False)
plt.legend(bbox_to_anchor=(1.05,1),loc=2)
plt.show()
```
